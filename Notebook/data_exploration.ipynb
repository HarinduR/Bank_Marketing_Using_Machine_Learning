# Handle Unknown Value


```python
import pandas as pd
```

```python
file_path = "/content/drive/MyDrive/MLCW/bank_additional_full_cleaned.csv"
data = pd.read_csv(file_path)
```

```python
data.head(10)
```

```python
print(data.info())
print(data.describe(include='all'))
```

```python
data.head(10)
```

```python
categorical_cols = ["job", "marital", "education", "default", "housing", "loan", "contact", "month", "day_of_week", "poutcome"]
for col in categorical_cols:
    print(f"Value counts for {col}:\n", data[col].value_counts())
```

```python
data.shape
```

```python
print(data.duplicated())
```

```python
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
```

```python
unknown_counts = data.isin(['unknown']).sum()
```

```python
print(unknown_counts)
```

```python
unknown_counts[unknown_counts > 0].plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title("Count of 'unknown' Values in Categorical Columns")
plt.xlabel("Columns")
plt.ylabel("Count")
plt.show()
```

```python
# Replace 'unknown' with 'other' in 'job' and 'education'
data['job'].replace('unknown', 'other', inplace=True)
data['education'].replace('unknown', 'other', inplace=True)
```

```python
# Replace 'unknown' with mode for 'marital', 'housing', and 'loan'
for col in ['marital', 'housing', 'loan']:
    mode_value = data[col].mode()[0]
    data[col].replace('unknown', mode_value, inplace=True)
```

```python
# Drop the 'default' column
data.drop('default', axis=1, inplace=True)
```

```python
data.drop('duration', axis=1, inplace=True)
```

```python
data.head(10)
```

```python
unknown_counts = data.isin(['unknown']).sum()
```

```python
print(unknown_counts)
```

```python
print(len(data))
```

# Handle Duplicate

```python
data.duplicated().sum()
```

```python
# Display duplicate rows
duplicate_rows = data[data.duplicated()]
print(duplicate_rows)
```

```python
# Remove duplicate rows
data = data.drop_duplicates()
print("Number of duplicates after removal:", data.duplicated().sum())
```

```python
print(len(data))
```
